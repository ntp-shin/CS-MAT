```python
import os
import shutil
```

```python
def copy_data(copy_path, paste_path):
    if not os.path.exists(paste_path):
        os.makedirs(paste_path)
    for i in range(10):
        shutil.copyfile(copy_path + str(i) + '.png', paste_path + str(i) + '.png')
```

```python
## Copy images
image_path = '../../data/CelebAMask-HQ/train/'
image_paste_path = './test_sets/CelebA-HQ/images/'
copy_data(image_path, image_paste_path)
```

```python
## Copy mask
mask_path = '../../data/CelebAMask-HQ/mask/'
mask_paste_path = './test_sets/CelebA-HQ/masks/'
copy_data(mask_path, mask_paste_path)
```

```python
# print height and width of images in folder
import os
import cv2

def get_image_size(path):
    count = 0
    for file in os.listdir(path):
        img = cv2.imread(path + '/' + file)
        height, width, channels = img.shape
        print(height, width, channels)
        if count > 10:
            return

# Convert size 1024x1024 to 512x512
def resize_images(path):
    count = 0
    for file in os.listdir(path):
        img = cv2.imread(path + '/' + file)
        img = cv2.resize(img, (512, 512))
        cv2.imwrite(path + '/' + file, img)
        count += 1
        if count > 10:
            return
```
